```python
# if ... else

# Ask the user for a number. Depending on whether the number is even or odd, print out an appropriate message to the user. 사용자에게 숫자를 물은 후 숫자가 짝수냐 홀수냐에 따라 적절한 메세지를 출력하세요

num = int(input('숫자를 입력해 주세요: '))

if num % 2 == 0:
    print('짝수를 입력하셨네요~')

else:
    print('홀수를 입력하셨네요~')
```

```python
# if .. elif .. else

# §Determine and print the letter grade corresponding to an input numeric grade. 숫자점수에 따라 문자점수를 결정하고 프린트하시오
# §Output
#   Enter the numeric grade: 88
#   The letter grade is B

score = int(input('점수를 입력해 주세요: '))

if score >= 90:
    print('A')

elif score >= 80:
    print('B')

elif score >= 70:
    print('C')

elif score >= 60:
    print('D')
```

```python
# Write a Python program to sum all the items in the list [1, 2, -8]. 리스트에 있는 숫자를 다 더하는 프로그램을 만드시오
lst = [1, 2, -8]
sum = 0
for l in lst:
    sum += l
print(sum)
```

```python
# Write a Python program to get the largest number from the list [1, 2, -8, 0]. 리스트에서 가장 큰 숫자를 찾는 프로그램을 만드시오
lst = [1, 2, -8, 0]
largest = lst[0]
for l in lst:
    if l >= largest:
        largest = l
print(largest)
```
