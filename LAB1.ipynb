# 세금 계산 프로그램

```python
print('세금 계산 프로그램')
print('='*20)
income = int(input('수입은? '))
num = int(input('부양가족은? '))

tax = (income - 10000 - num * 3000)*.2
print('세금은', int(tax), '입니다.')
```
