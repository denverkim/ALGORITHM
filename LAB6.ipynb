```python
import pandas as pd
import numpy as np
```

```python
data = pd.Series([1, np.nan, 'hello', None])
data
```

```python
# 널값이 있는 곳에 TRUE 리턴
data.isna()
data.isnull()
```

```python
# 널값 요약
print(data.isna().sum())
```

```python
# 널값이 아닌 데이터 저장
data1 = data[data.notnull()]
data1
```

```python
df = pd.DataFrame([[1, np.nan, 2], [2, 3, 5], [np.nan, 4, 6], [np.nan, np.nan, np.nan]])
df.columns = ['c1', 'c2', 'c3']
df.isna().sum()
df.shape
df.mean()
df
```

```python
# 전체 행이 널값인 경우 제거
df.dropna(how='all')
```

```python
# 널값이 하나라도 있는 열 제거
df.dropna(axis=1)
```

```python
df.dropna(how='all')
```

```python
df.fillna(0)
```
