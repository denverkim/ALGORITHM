```python
# Read weather.txt file into a dataframe.

import pandas as pd

weather = pd.read_csv('/content/weather.txt')
weather
```

```python
# Print first 5 or last 3 rows of df 처음 다섯줄, 마지막 세줄 출력
print('상위 5개 데이터 출력')
print(weather.head())

print('\n하위 3개 데이터 출력')
print(weather.tail(3))

# Get data types, index, columns, values 데이터타입, 인덱스, 컬럼들, 값들
print('\n컬럼별 데이터 타입')
print(weather.dtypes)

# Statistical summary of each column 각 컬럼에 대한 통계요약
print('\n기초 통계량')
print(weather.describe())

# Sort records by any column (descending order)
# 한 컬럼을 정해 내림차순으로 정렬
weather.sort_values('avg_precipitation', ascending=False)
```

```python
weather1 = weather.drop(['month'], axis=1)
print('\n평균')
print(weather1.mean(axis=1))
print('\n중위수')
print(weather1.median())
print('\n합계')
print(weather1.sum())
print('\n최소값')
print(weather1.min())
print('\n최대값')
print(weather1.max())
print('\n분산')
print(weather1.var())
```

```python
# Slice the records and display the following columns and rows:
# 레코드를 짤라서 다음의 열과 행을 출력
# avg_low
weather.columns
weather.avg_low
weather['avg_low']

# Rows 1 to 2
weather[:2]

# avg_low and avg_high
weather[['avg_low', 'avg_high']]

# 9 row of avg_precipitation column
weather.loc[8, 'avg_precipitation']

# 4 to 5 rows of 1 and 4 columns
weather.iloc[3:5, [0,3]]
```

```python
# Filter the data and display the following columns and rows:
# 데이터를 검색하여 다음의 열과 행을 출력
# avg_precipitation > 1.0

weather_1 = weather[weather.avg_precipitation > 1.0]

# Month is in either June, July, or August
weather_2 = weather[weather.month.isin(['Jun', 'Jul', 'Aug'])]
```

```python
# Assign new values in the following locations:
# 101.3 for avg_precipitation column at index 9
weather.loc[9, 'avg_precipitation'] = 101.3
weather

# Np null values for avg_precipitation column at index 9 (np.nan)
import numpy as np

weather.loc[9, 'avg_precipitation'] = np.nan
weather

# 5 for all rows in avg_low column avg_low 컬럼 전체를 5로 바꾸기

weather.avg_low = 5
weather

# Add new column named avg_day that is the average of avg_low and avg_high avg_low와 avg_high의 평균을 구하여 avg_day컬럼을 만듬
weather['avg_day'] = (weather.avg_low + weather.avg_high)/2
weather
```

```python
# Rename columns 컬럼이름 바꾸기
# avg_precipitation to avg_rain
weather.rename(columns={'avg_precipitation': 'avg_rain'}, inplace=True)

# Change columns’ name to 'month','av_hi','av_lo','rec_hi','rec_lo','av_rain‘, ‘av_day’
weather.columns = ['month','av_hi','av_lo','rec_hi','rec_lo','av_rain', 'av_day']

# Save the result data frame to a csv file 데이터플레임을 csv파일로 저장
weather.to_csv('/content/weather.csv')
```
