```python
import pandas as pd
import matplotlib.pyplot as plt

chipo = pd.read_excel('/content/chipotle.xlsx')
chipo.columns
x = chipo.item_price
y = chipo.quantity

# Scatter Plot
plt.scatter(x, y, color='k', s=100)
plt.xlabel('Item Price')
plt.ylabel('Quantity')
plt.title('Scatter Plot')
plt.show()
```

```python
# Histogram
plt.hist(chipo.item_price, rwidth=.8)
plt.xlabel('Item Price')
plt.ylabel('Count')
plt.title('Histogram')
plt.show()
```

```python
# Bar chart
freq = chipo.item_name.value_counts()
x = freq.index
y = freq.values

plt.bar(x, y)
plt.xlabel('Item Price')
plt.ylabel('Count')
plt.title('Bar Chart')
plt.xticks(rotation=90)
plt.show()
```

```python
# Horizontal bar chart
plt.barh(x, y, color='r')
plt.xlabel('Item Name')
plt.ylabel('Count')
plt.title('Horizontal Bar Chart')
plt.show()
```

```python
# Heat map
plt.pcolor(chipo[['item_price', 'quantity']].corr())
plt.colorbar()
plt.title('Heat Map')
plt.show()
```
